## Correlação

```python
corr = df.corr(numeric_only=True)

corr["distancia"].sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,5))
corr["distancia"].sort_values().plot(kind="barh")
plt.title("Correlação com Distância")
plt.tight_layout()
plt.show()
```
